> Import library yang diperlukan


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
```

> Mengambil File Image dari GitHub

```python
url = 'https://raw.githubusercontent.com/jurus231/Image-Processing-Point-Operation-/main/mukapikri.jpg'
```

```python
resp = urllib.request.urlopen(url)
image_array = np.asarray(bytearray(resp.read()), dtype=np.uint8)
```



> Membaca Image yang telah diambil



```python
image_color = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.title("Muka Pikri")
plt.show()
```

> Melakukan penambahan kecerahan pada image

```python
bright_image = cv2.add(image_rgb, 50)
plt.imshow(bright_image)
plt.axis('off')
plt.title("Muka Pikri tapi Cerah")
plt.show()
```

> Operasi Inversi Negatif Pada Image

```python
# Inversi warna untuk setiap channel (R, G, B)
negative_color = 255 - image_color

# Convert kembali ke RGB untuk tampilan yang benar
negative_rgb = cv2.cvtColor(negative_color, cv2.COLOR_BGR2RGB)

# Tampilkan hasilnya
plt.imshow(negative_rgb)
plt.axis('off')
plt.title("Inversi (Negatif)")
plt.show()
```

> Operasi Thresholding pada Image

```python
# Convert gambar menjadi grayscale terlebih dahulu
gray_image = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)

# Lakukan threshold pada gambar grayscale
_, threshold_color = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Convert kembali threshold ke gambar berwarna
threshold_bgr = cv2.cvtColor(threshold_color, cv2.COLOR_GRAY2BGR)
threshold_rgb = cv2.cvtColor(threshold_bgr, cv2.COLOR_BGR2RGB)

# Tampilkan hasilnya
plt.imshow(threshold_rgb)
plt.axis('off')
plt.title("Thresholding")
plt.show()
```

> Operasi penambahan Kontras pada Image

```python
# Perbesar kontras untuk setiap channel (R, G, B)
contrast_color = np.clip(image_color * 1.5, 0, 255).astype(np.uint8)

# Convert kembali ke RGB untuk tampilan yang benar
contrast_rgb = cv2.cvtColor(contrast_color, cv2.COLOR_BGR2RGB)

# Tampilkan hasilnya
plt.imshow(contrast_rgb)
plt.axis('off')
plt.title("Kontras 1.5")
plt.show()
```
